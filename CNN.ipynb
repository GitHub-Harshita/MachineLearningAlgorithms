```python
import numpy as np
X = np.random.rand(100,64,64,3) #100 image with 64*64 pixels and 3 is 3rcb channel
y = np.random.randint(0,2,100) #0 means cat and 1 means dog
```

isme humnr matrix isiliye li hai kyuki cnn ke input me image matrix ki form me hi jati h, and even jo images humari system me saved hoti h vo b matrix ki form me hi hoti hai, pixels(rgb) fformat me combine krke like rgb [0 0 255]

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(64,64,3)),   #ye basic feautures phchan ne ke liye jse edge, vertices
    MaxPooling2D(2,2),   #image size chota krta hai, taki overfitting na ho next layer pr
    Conv2D(64, (3,3), activation='relu'), #layers bdhadi, ab ye deep feautures analyse krega, jse shape and pattern, faces
    MaxPooling2D(2,2),   # fr se size chota hoga, overfitting na ho isiliye
    Flatten(),          #2d vector ko 1d me convert krega
    Dense(128, activation='relu'),  #sequential model ki hidden layer se pass kraya
    Dense(1, activation='sigmoid')  #output layer
])
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=10)
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img("dog.jpg", target_size=(64,64))
```

```python
img_array=image.img_to_array(img)  #Load the image, output iska (width, height, channel) hota h, means (64,64,3)
```

```python
import numpy as np
```

```python
img_array=np.expand_dims(img_array, axis=0) #dimensions expand kre kyuki cnn imput me batch size b leta h, chahe 1 image hi ho
```

```python
prediction=model.predict(img_array)
```
