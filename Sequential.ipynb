SEQUENTIAL MODEL

```python
import numpy as np
```

```python
X = np.array([
    [2, 5],
    [4, 6],
    [6, 7],
    [8, 8]])
y = np.array([40, 60, 75, 90])
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(8, activation="relu",input_shape=([2])),
    Dense(1)
])
```

```python
model.compile(optimizer="adam", loss="mse")
```

```python
model.fit(X,y,epochs=300)
```

```python
model.predict(np.array([[7,8]]))
```

Ye jo 20 predict kiya h isne ye galat kiya hai kyuki model humara uderfit h, mtlb ki model hmara bhaut simple h ki isne pdhna hi strt nhi kiya h

KYUKI YE CONFUSE H KI DATA KO konsi range me lu x vle data ki ya y vle data ki, to hume isko scale krna pdega 0 se 1 ke bich me, taki ye ache se sikh paye

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Scaler_X=MinMaxScaler()
Scaler_y=MinMaxScaler()
```

```python
scaled_X=Scaler_X.fit_transform(X)
```

```python
scaled_y=Scaler_y.fit_transform(y.reshape(-1,1)) #reshape isiliye kiya kyuki keras ko 2d data chahiye hota h or y hmara 1d h, isme -1 ka mtlb ye khud calculate krega no. of rows or 1 mtlb 1 column
```

```python
model.fit(scaled_X, scaled_y, epochs=300)
```

```python
pred=model.predict(Scaler_X.transform([[7,8]]))
```

```python
pred
```

```python
pred_original=Scaler_y.inverse_transform(pred)
```

```python
pred_original
```
